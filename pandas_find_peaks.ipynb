```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Sample data
data = {
    'values': [1, 3, 7, 1, 2, 6, 5, 2, 8, 4]
}
df = pd.DataFrame(data)

# Find local maxima
peaks, _ = find_peaks(df['values'])

# Find local minima
# To find minima, we invert the values and use find_peaks
inverted_peaks, _ = find_peaks(-df['values'])

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['values'], label='Data', marker='o')

# Plot local maxima
plt.plot(peaks, df['values'].iloc[peaks], 'ro', markersize=5, label='Local Maxima')

# Plot local minima
plt.plot(inverted_peaks, df['values'].iloc[inverted_peaks], 'bo', markersize=5, label='Local Minima')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Local Maxima and Minima')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
